Q1 Dataset Exploration

```python
import pandas as pd

df = pd.read_csv("Advanced_pandas_sales_dataset.csv")

# First 10 rows
print(df.head(10))

# Dataset information
print(df.info())

# Rows and Columns
print("Rows, Columns:", df.shape)

# Data types
print(df.dtypes)
```

Q2 Sales Filtering Analysis

```python
# Sales > 3500
sales_above_3500 = df[df["Sales"] > 3500]

# Region = East and Sales > 2500
east_sales = df[(df["Region"] == "East") & (df["Sales"] > 2500)]

print(sales_above_3500)
print(east_sales)
```

Q3 Customer Purchase Analysis

```python
top_customers = (
    df.groupby("Customer")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

print(top_customers)
```

Q4 Regional Sales Performance

```python
region_sales = df.groupby("Region")["Sales"].sum()
print(region_sales)
```

Q5 Product Performance Analysis

```python
product_qty = (
    df.groupby("Product")["Quantity"]
      .sum()
      .sort_values(ascending=False)
)

print(product_qty)
```

Q6 Multi-Level Grouping

```python
region_category_sales = (
    df.groupby(["Region", "Category"])["Sales"]
      .sum()
)

print(region_category_sales)
```

Q7 Pivot Table Creation

```python
pivot_table = pd.pivot_table(
    df,
    index="Region",
    columns="Product",
    values="Sales",
    aggfunc="sum",
    fill_value=0
)

print(pivot_table)
```

Q8 Profitability Analysis

```python
avg_profit = (
    df.groupby("Category")["Profit"]
      .mean()
      .sort_values(ascending=False)
)

print(avg_profit)
```

Q9 Discount Impact Study

```python
discount_profit = (
    df.groupby("Discount")["Profit"]
      .mean()
)

print(discount_profit)
```

Q10 End-to-End Data Wrangling Task

```python
filtered_df = df[df["Sales"] > 3000]

result = (
    filtered_df
    .groupby(["Region", "Product"])["Sales"]
    .sum()
    .reset_index()
)

pivot_result = result.pivot(
    index="Region",
    columns="Product",
    values="Sales"
).fillna(0)

print(pivot_result)
```
